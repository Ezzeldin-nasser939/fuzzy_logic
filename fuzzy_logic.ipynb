```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

#Cholesterol
Cholesterol = ctrl.Antecedent(np.arange(0, 501, 1), 'Cholesterol')
Cholesterol['Normal'] = fuzz.trimf(Cholesterol.universe, [0, 0, 200])
Cholesterol['Medium'] = fuzz.trimf(Cholesterol.universe, [190,250 , 250])
Cholesterol['High'] = fuzz.trimf(Cholesterol.universe, [230, 320, 320])
Cholesterol['Very High'] = fuzz.trimf(Cholesterol.universe, [280,500 , 500])

#Blood Pressure
Blood_Pressure = ctrl.Antecedent(np.arange(0, 201, 1), 'Blood Pressure')
Blood_Pressure['Normal'] = fuzz.trimf(Blood_Pressure.universe, [0, 0, 130])
Blood_Pressure['Medium'] = fuzz.trimf(Blood_Pressure.universe, [120,159 , 159])
Blood_Pressure['Very High'] = fuzz.trimf(Blood_Pressure.universe, [150, 200, 200])

#Physical Activity
Physical_Activity = ctrl.Antecedent(np.arange(0, 1.1, .1), 'Physical Activity')
Physical_Activity['False'] = fuzz.trimf(Physical_Activity.universe, [0, 0, .6])
Physical_Activity['True'] = fuzz.trimf(Physical_Activity.universe, [.3, 1, 1])

#Age
Age  = ctrl.Antecedent(np.arange(0, 76, 1), 'Age')
Age['Young'] = fuzz.trimf(Age.universe, [0, 0, 38])
Age['Middle age'] = fuzz.trimf(Age.universe, [34,45,45])
Age['Old'] = fuzz.trimf(Age.universe, [40, 58, 58])
Age['Very old'] = fuzz.trimf(Age.universe, [53,75 , 75])

#BMI
BMI = ctrl.Antecedent(np.arange(0, 51, 1), 'BMI')
BMI['Normal'] = fuzz.trimf(BMI.universe, [0, 0, 25])
BMI['Over weight'] = fuzz.trimf(BMI.universe, [24,32 , 32])
BMI['Obese'] = fuzz.trimf(BMI.universe, [30, 50, 50])

#Smoking
Smoking =ctrl.Antecedent(np.arange(0, 11, 1), 'Smoking')
Smoking['False'] = fuzz.trimf(Smoking.universe, [0, 0, 6])
Smoking['True'] = fuzz.trimf(Smoking.universe, [3,10,10])

#Diabetes
Diabetes =ctrl.Antecedent(np.arange(0, 401, 1), 'Diabetes')
Diabetes['Normal'] = fuzz.trimf(Diabetes.universe, [0, 0, 160])
Diabetes['Diabetic'] = fuzz.trimf(Diabetes.universe, [150, 400, 400])

#indicates heart disease
indicates_heart_disease = ctrl.Consequent(np.arange(0, 3.1, .1), 'indicates heart disease')
indicates_heart_disease['Healthy'] = fuzz.trimf(indicates_heart_disease.universe, [0.0, 0.0, 1.5])
indicates_heart_disease['Early stage'] = fuzz.trimf(indicates_heart_disease.universe, [1.3,2.5 , 2.5])
indicates_heart_disease['Advaned stage'] = fuzz.trimf(indicates_heart_disease.universe, [2.2,3.0 , 3.0])
```

```python
Age.view()
Blood_Pressure.view()
Cholesterol.view()
Diabetes.view()
Smoking.view()
BMI.view()
indicates_heart_disease.view()
```

```python
rule1 = ctrl.Rule(Age['Young'] & Blood_Pressure['Normal'] & Cholesterol['Normal'] & Diabetes['Diabetic'] & BMI['Normal'] & Physical_Activity['True'] & Smoking['False'],indicates_heart_disease['Healthy'])
rule2 = ctrl.Rule(Age['Middle age'] & Blood_Pressure['Normal'] & Cholesterol['Normal'] & Diabetes['Diabetic'] & BMI['Normal'] & Physical_Activity['True'] & Smoking['False'],indicates_heart_disease['Healthy'])
rule3 = ctrl.Rule(Age['Old'] & Blood_Pressure['Normal'] & Cholesterol['Normal'] & Diabetes['Diabetic'] & BMI['Normal'] & Physical_Activity['True'] & Smoking['True'],indicates_heart_disease['Healthy'])
rule4 = ctrl.Rule(Age['Very old'] & Blood_Pressure['Normal'] & Cholesterol['Normal'] & Diabetes['Diabetic'] & BMI['Normal'] & Physical_Activity['True'] & Smoking['False'],indicates_heart_disease['Healthy'])
rule5 = ctrl.Rule(Age['Young'] & Blood_Pressure['Medium'] & Cholesterol['Medium'] & Diabetes['Normal'] & BMI['Over weight'] & Physical_Activity['False'],indicates_heart_disease['Healthy'])
rule6 = ctrl.Rule(Age['Middle age'] & Blood_Pressure['Medium'] & Cholesterol['Medium'] & Diabetes['Normal'] & BMI['Over weight'] & Physical_Activity['False'] ,indicates_heart_disease['Healthy'])
rule7 = ctrl.Rule(Age['Old'] & Blood_Pressure['Medium'] & Cholesterol['Medium'] & Diabetes['Normal'] & BMI['Over weight'] & Physical_Activity['True'] & Smoking['False'],indicates_heart_disease['Healthy'])
rule8 = ctrl.Rule(Age['Very old'] & Blood_Pressure['Medium'] & Cholesterol['Medium'] & Diabetes['Normal'] & BMI['Over weight'] & Physical_Activity['True'] & Smoking['False'],indicates_heart_disease['Healthy'])
rule9 = ctrl.Rule(Age['Old'] & Blood_Pressure['Very High'] & Cholesterol['High'] & Diabetes['Diabetic'] & BMI['Normal'] & Physical_Activity['True'] & Smoking['False'],indicates_heart_disease['Healthy'])
rule10 = ctrl.Rule(Age['Very old'] & Blood_Pressure['Very High'] & Cholesterol['Normal'] & Diabetes['Diabetic'] & BMI['Obese'] & Physical_Activity['False'] & Smoking['False'],indicates_heart_disease['Healthy'])
rule11 = ctrl.Rule(Age['Young'] & Blood_Pressure['Very High'] & Cholesterol['High'] & Diabetes['Diabetic'] & BMI['Obese'] & Physical_Activity['False'] & Smoking['True'],indicates_heart_disease['Early stage'])
rule12 = ctrl.Rule(Age['Middle age'] & Blood_Pressure['Very High'] & Cholesterol['High'] & Diabetes['Normal'] & BMI['Obese'] & Physical_Activity['False'] & Smoking['True'],indicates_heart_disease['Early stage'])
rule13 = ctrl.Rule(Age['Old'] & Blood_Pressure['Very High'] & Cholesterol['High'] & Diabetes['Normal'] & BMI['Obese'] & Physical_Activity['False'] & Smoking['True'],indicates_heart_disease['Early stage'])
rule14 = ctrl.Rule(Age['Very old'] & Blood_Pressure['Very High'] & Cholesterol['High'] & Diabetes['Normal'] & BMI['Obese'] & Physical_Activity['False'] & Smoking['True'],indicates_heart_disease['Early stage'])
rule15 = ctrl.Rule(Age['Very old'] & Blood_Pressure['Normal'] & Cholesterol['Very High'] & Diabetes['Normal'] & BMI['Over weight'] & Physical_Activity['True'] ,indicates_heart_disease['Early stage'])
rule16 = ctrl.Rule(Age['Very old'] & Blood_Pressure['Medium'] & Cholesterol['Very High'] & Diabetes['Normal'] & BMI['Over weight'] & Physical_Activity['True'] ,indicates_heart_disease['Early stage'])
rule17 = ctrl.Rule(Age['Very old'] & Blood_Pressure['Very High'] & Cholesterol['Very High'] & Diabetes['Normal'] & BMI['Obese'] & Physical_Activity['False']& Smoking['True'] ,indicates_heart_disease['Advaned stage'])
rule18 = ctrl.Rule(Age['Old'] & Blood_Pressure['Very High'] & Cholesterol['Very High'] & Diabetes['Normal'] & BMI['Obese'] & Physical_Activity['False'],indicates_heart_disease['Advaned stage'])
rule19 = ctrl.Rule(Age['Old'] & Blood_Pressure['Very High'] & Cholesterol['Very High'] & Diabetes['Normal'] & BMI['Obese'] & Physical_Activity['False']& Smoking['True'],indicates_heart_disease['Advaned stage'])
rule20 = ctrl.Rule(Age['Old'] & Blood_Pressure['Very High'] & Cholesterol['Very High'] & Diabetes['Normal'] & BMI['Obese'] & Physical_Activity['True']& Smoking['False'],indicates_heart_disease['Advaned stage'])
rule21 = ctrl.Rule(Age['Very old'] & Blood_Pressure['Very High'] & Cholesterol['Very High'] & Diabetes['Normal'] & BMI['Normal'] & Physical_Activity['True']& Smoking['False'],indicates_heart_disease['Advaned stage'])
rule22 = ctrl.Rule(Age['Very old'] | Blood_Pressure['Very High'] & Cholesterol['Very High'] | Diabetes['Normal'] & BMI['Normal'] & Physical_Activity['True'] | Smoking['False'],indicates_heart_disease['Advaned stage'])
rule23 = ctrl.Rule(Age['Old'] | Blood_Pressure['Very High'] & Cholesterol['Very High'] | Diabetes['Normal'] & BMI['Normal'] & Physical_Activity['True'] | Smoking['True'],indicates_heart_disease['Advaned stage'])
rule24 = ctrl.Rule(Age['Old'] | Blood_Pressure['Very High'] & Cholesterol['High'] | Diabetes['Normal'] & BMI['Normal'] & Physical_Activity['True'] | Smoking['False'],indicates_heart_disease['Early stage'])
rule25 = ctrl.Rule(Age['Young'] | Blood_Pressure['Medium'] & Cholesterol['Medium'] | Diabetes['Normal'] & BMI['Normal'] & Physical_Activity['True'] | Smoking['False'],indicates_heart_disease['Healthy'])

rule1.view()
```

```python
indicates_ctrl= ctrl.ControlSystem([rule1, rule2, rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25])
indicates = ctrl.ControlSystemSimulation(indicates_ctrl)
```
